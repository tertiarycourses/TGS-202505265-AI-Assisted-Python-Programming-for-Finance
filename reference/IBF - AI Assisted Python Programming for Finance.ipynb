## Activity: CPF Calculator

```python
def calculate_cpf(salary, age):
    # Basic CPF Contribution Rates (Employee and Employer)
    if age <= 55:
        emp_rate, emr_rate = 0.20, 0.17
    elif age <= 60:
        emp_rate, emr_rate = 0.15, 0.145
    elif age <= 65:
        emp_rate, emr_rate = 0.095, 0.11
    elif age <= 70:
        emp_rate, emr_rate = 0.07, 0.09
    else:
        emp_rate, emr_rate = 0.05, 0.075

    # Salary ceiling is $6,800 as of 2024
    capped_salary = min(salary, 6800)

    employee_contribution = capped_salary * emp_rate
    employer_contribution = capped_salary * emr_rate
    total_cpf = employee_contribution + employer_contribution

    return employee_contribution, employer_contribution, total_cpf

# Input
sal = float(input("Enter monthly salary: "))
age = int(input("Enter age: "))

emp, emr, total = calculate_cpf(sal, age)

print(f"\nEmployee Contribution: ${emp:.2f}")
print(f"Employer Contribution: ${emr:.2f}")
print(f"Total CPF Contribution: ${total:.2f}")
```

### Activity: Return of Equity (ROE)

```python
import yfinance as yf
import pandas as pd

# DBS Group Holdings Ticker on SGX
dbs_ticker = 'D05.SI'
dbs = yf.Ticker(dbs_ticker)

# Fetch Income Statement and Balance Sheet
income_stmt = dbs.financials
balance_sheet = dbs.balance_sheet

try:
    # Get Net Income (latest annual)
    net_income = income_stmt.loc['Net Income'].iloc[0]

    # Get Stockholders Equity (latest and previous for average)
    equity_latest = balance_sheet.loc['Stockholders Equity'].iloc[0]
    equity_prev = balance_sheet.loc['Stockholders Equity'].iloc[1]
    avg_equity = (equity_latest + equity_prev) / 2

    roe = (net_income / avg_equity) * 100

    print(f"COMPANY: DBS GROUP HOLDINGS ({dbs_ticker})")
    print(f"NET INCOME: ${net_income:,.0f}")
    print(f"AVERAGE SHAREHOLDERS EQUITY: ${avg_equity:,.0f}")
    print(f"RETURN ON EQUITY (ROE): {roe:.2f}%")

    # Recommendation logic based on a threshold (e.g., 15%)
    threshold = 15.0
    if roe > threshold:
        recommendation = "BUY (High ROE)"
    else:
        recommendation = "SELL (Low ROE)"

    print(f"\nRECOMMENDATION: {recommendation} (Threshold: {threshold}%)")

except Exception as e:
    print(f"ERROR EXTRACTING DATA: {e}")
```

### Activity: Fibonanaci Sequence

```python
n_terms = int(input("How many terms? "))

a, b = 0, 1
count = 0

fibonacci_sequence = []
ratios = []

print(f"\nCalculating Fibonacci sequence for {n_terms} terms:")

while count < n_terms:
    fibonacci_sequence.append(a)

    # Calculate ratio of b/a to show Golden Ratio convergence (skip first few terms to avoid div by zero)
    if a != 0:
        ratio = b / a
        ratios.append(ratio)

    # Multiple assignment for next terms
    a, b = b, a + b
    # a = b
    # b = a+b
    count += 1

print(f"Sequence: {fibonacci_sequence}")

if len(ratios) > 0:
    print(f"\nConvergence towards Golden Ratio (Last 5 ratios):")
    for r in ratios[-5:]:
        print(f"{r:.6f}")
    print(f"Target Golden Ratio: ~1.618034")
```

## Activity: Return of Asset (ROA)

```python
import yfinance as yf
import pandas as pd

# List of 10 major SGX tickers
sgx_tickers = [
    'D05.SI', 'U11.SI', 'O39.SI', 'Z74.SI', 'J36.SI',
    'BN4.SI', 'V03.SI', 'C38U.SI', 'A17U.SI', 'C07.SI'
]

roa_results = []

print("Fetching data and calculating ROA...")

for ticker in sgx_tickers:
    try:
        stock = yf.Ticker(ticker)
        # Fetching Net Income from Income Statement
        income_stmt = stock.financials
        # Fetching Total Assets from Balance Sheet
        balance_sheet = stock.balance_sheet

        # ROA = Net Income / Total Assets
        net_income = income_stmt.loc['Net Income'].iloc[0]
        total_assets = balance_sheet.loc['Total Assets'].iloc[0]

        roa = (net_income / total_assets) * 100
        name = stock.info.get('shortName', ticker)

        roa_results.append({"name": name, "ticker": ticker, "roa": roa})
    except Exception as e:
        print(f"Could not process {ticker}: {e}")

# Rank from highest to lowest ROA
ranked_roa = sorted(roa_results, key=lambda x: x['roa'], reverse=True)

print("\nSingapore Companies Ranked by ROA:")
print(f"{'Rank':<5} {'Company':<25} {'Ticker':<10} {'ROA (%)':<10}")
for i, data in enumerate(ranked_roa, 1):
    print(f"{i:<5} {data['name']:<25} {data['ticker']:<10} {data['roa']:>8.2f}%")
```

### Activity: List Comprehension

```python
import yfinance as yf

# List of 10 major SGX tickers
sgx_tickers = ['D05.SI', 'U11.SI', 'O39.SI', 'Z74.SI', 'J36.SI', 'BN4.SI', 'V03.SI', 'C38U.SI', 'A17U.SI', 'C07.SI']

print("Fetching financial data and calculating Net Profit Margin...")

# Helper function to get margin safely
def get_margin(ticker):
    try:
        stock = yf.Ticker(ticker)
        financials = stock.financials
        net_income = financials.loc['Net Income'].iloc[0]
        revenue = financials.loc['Total Revenue'].iloc[0]
        return {'ticker': ticker, 'margin': (net_income / revenue) * 100}
    except:
        return {'ticker': ticker, 'margin': None}

# Using List Comprehension to compute margins
results = [get_margin(t) for t in sgx_tickers]

print(f"\n{'Ticker':<10} {'Net Profit Margin (%)':<20}")
print("-" * 30)
for res in results:
    # Correcting 'null' to 'None' for Python syntax
    margin_str = f"{res['margin']:.2f}%" if res['margin'] is not None else "N/A"
    print(f"{res['ticker']:<10} {margin_str:<20}")
```

# Topic 4 Scripting with Function and Lambda

## Functions

```python
def calculate_cagr(beginning_value, ending_value, years):
    """Calculates the Compound Annual Growth Rate."""
    return ((ending_value / beginning_value) ** (1 / years)) - 1

def calculate_simple_interest(principal, rate, time):
    """Calculates simple interest: I = P * r * t"""
    return principal * (rate / 100) * time

def break_even_units(fixed_costs, price_per_unit, variable_cost_per_unit):
    """Calculates number of units to break even."""
    return fixed_costs / (price_per_unit - variable_cost_per_unit)

# Example Usage:
cagr = calculate_cagr(10000, 15000, 5)
interest = calculate_simple_interest(5000, 4.5, 2)
units = break_even_units(20000, 50, 30)

print(f"CAGR: {cagr:.2%}")
print(f"Simple Interest: ${interest:.2f}")
print(f"Break-even Units: {units} units")
```

### Return Multiple Values

```python
def analyze_income_statement(revenue, cogs, operating_expenses):
    """
    Returns Gross Profit, Net Income, and Profit Margin as a tuple.
    """
    gross_profit = revenue - cogs
    net_income = gross_profit - operating_expenses
    profit_margin = (net_income / revenue) * 100 if revenue != 0 else 0

    return gross_profit, net_income, profit_margin

# Example usage with sample data
rev = 500000
cost = 200000
op_ex = 150000

# Unpacking multiple returned values
gp, ni, pm = analyze_income_statement(rev, cost, op_ex)

print(f"Revenue: ${rev:,}")
print(f"Gross Profit: ${gp:,}")
print(f"Net Income: ${ni:,}")
print(f"Profit Margin: {pm:.2f}%")
```

```python
def calculate_net_profit_margin(net_income, revenue):
    """
    Computes the Net Profit Margin as a percentage.
    Formula: (Net Income / Total Revenue) * 100
    """
    if revenue == 0:
        return 0
    return (net_income / revenue) * 100

# Example usage with sample data:
sample_net_income = 1000000
sample_revenue = 5000000

margin = calculate_net_profit_margin(sample_net_income, sample_revenue)
print(f"Calculated Net Profit Margin: {margin:.2f}%")
```

### Default Arguments

```python
def calculate_net_salary(gross_salary, tax_rate=0.15):
    """
    Calculates net salary after tax.
    Default tax rate is 15% (0.15).
    """
    tax_amount = gross_salary * tax_rate
    return gross_salary - tax_amount

# Example 1: Using the default tax rate
sal_default = calculate_net_salary(5000)

# Example 2: Providing a custom tax rate (e.g., 20%)
sal_custom = calculate_net_salary(5000, tax_rate=0.20)

print(f"Net Salary (Default 15% Tax): ${sal_default:,.2f}")
print(f"Net Salary (Custom 20% Tax): ${sal_custom:,.2f}")

def estimate_investment_value(principal, years, annual_return=0.07):
    """
    Estimates future value of an investment.
    Default annual return is 7% (0.07).
    """
    return principal * (1 + annual_return) ** years

# Example 3: Using the default 7% return
future_val = estimate_investment_value(1000, 5)
print(f"\nInvestment Value after 5 years (7% return): ${future_val:,.2f}")
```

### Named Arguments

```python
def calculate_dividend_yield(dividend_per_share, stock_price):
    """Calculates dividend yield."""
    return (dividend_per_share / stock_price) * 100

# Using Named Arguments (Keyword Arguments)
# This makes it clear which value corresponds to which parameter
yield_a = calculate_dividend_yield(dividend_per_share=2.50, stock_price=50.00)

# With named arguments, the order doesn't matter
yield_b = calculate_dividend_yield(stock_price=120.00, dividend_per_share=3.60)

print(f"Dividend Yield A: {yield_a:.2f}%")
print(f"Dividend Yield B: {yield_b:.2f}%")

def calculate_roi(initial_investment, final_value):
    """Calculates Return on Investment."""
    return ((final_value - initial_investment) / initial_investment) * 100

# Example of clarity using named arguments
my_roi = calculate_roi(initial_investment=1000, final_value=1250)
print(f"\nROI: {my_roi:.2f}%")
```

### Variable Arguments

```python
def calculate_portfolio_total(*stock_values):
    """
    Accepts any number of stock values and returns the total sum.
    """
    total = sum(stock_values)
    return total

# Example 1: Calculating total value for a portfolio of 3 stocks
portfolio_a = calculate_portfolio_total(1500.50, 2300.75, 850.00)

# Example 2: Calculating total value for a portfolio of 5 stocks
portfolio_b = calculate_portfolio_total(500, 1200, 450, 3000, 150)

print(f"Total Value Portfolio A (3 stocks): ${portfolio_a:,.2f}")
print(f"Total Value Portfolio B (5 stocks): ${portfolio_b:,.2f}")

def list_expenses(category, *amounts):
    """
    Accepts a category name and a variable number of expense amounts.
    """
    total = sum(amounts)
    print(f"\nCategory: {category}")
    print(f"Number of items: {len(amounts)}")
    print(f"Total Expenses: ${total:,.2f}")

# Example 3: Different numbers of arguments
list_expenses("Marketing", 120.50, 450.00, 89.90)
list_expenses("Utilities", 300.25)
```

### Activity: Max ROE

```python
import yfinance as yf

# List of 10 major SGX tickers
sgx_tickers = ['D05.SI', 'U11.SI', 'O39.SI', 'Z74.SI', 'J36.SI', 'BN4.SI', 'V03.SI', 'C38U.SI', 'A17U.SI', 'C07.SI']

roe_list = []

print("Fetching ROE data for 10 Singapore companies...")
for ticker in sgx_tickers:
    try:
        stock = yf.Ticker(ticker)
        # returnOnEquity is usually returned as a decimal (e.g., 0.15 for 15%)
        roe = stock.info.get('returnOnEquity')
        if roe is not None:
            roe_list.append({'ticker': ticker, 'roe': roe * 100})
    except:
        continue

def find_max_roe(data_list):
    """
    Finds the maximum ROE in a list of dictionaries.
    """
    if not data_list:
        return None

    max_entry = data_list[0]
    for entry in data_list:
        if entry['roe'] > max_entry['roe']:
            max_entry = entry
    return max_entry

# Display the sequence
print("\nROE Sequence:")
for item in roe_list:
    print(f"{item['ticker']}: {item['roe']:.2f}%")

# Find and print the max
highest = find_max_roe(roe_list)
if highest:
    print(f"\nThe maximum ROE is {highest['roe']:.2f}% belonging to {highest['ticker']}.")
```

### Activity: ROE Map

```python
# Hardcoded financial data for 5 SGX companies (Net Income and Shareholders' Equity)
# Values are in SGD billions for demonstration
financial_data = [
    {'ticker': 'D05.SI', 'net_income': 10.3, 'equity': 57.1},
    {'ticker': 'U11.SI', 'net_income': 5.7, 'equity': 45.2},
    {'ticker': 'O39.SI', 'net_income': 7.0, 'equity': 54.8},
    {'ticker': 'Z74.SI', 'net_income': 1.9, 'equity': 26.5},
    {'ticker': 'V03.SI', 'net_income': 1.1, 'equity': 12.4}
]

# Lambda function to compute ROE (Net Income / Equity * 100)
roe_calculator = lambda x: {
    'ticker': x['ticker'],
    'roe': (x['net_income'] / x['equity']) * 100 if x['equity'] != 0 else 0
}

# Using map function to apply the lambda across the hardcoded list
roe_results = list(map(roe_calculator, financial_data))

print(f"{'Ticker':<10} {'Calculated ROE (%)':<20}")
print("-" * 30)
for res in roe_results:
    print(f"{res['ticker']:<10} {res['roe']:>10.2f}%")
```
